```python
import pandas as pd
import sqlite3

#Aqui agrega la conexión a la base de datos
conn = sqlite3.connect("Salaries.sqlite")

#Crea el dataframe con los datos del datasets
df = pd.read_sql_query("SELECT * FROM Salaries", conn)
df.head()
```

1. **Top 10 empleados con mayor salario total (incluyendo beneficios)**
Extrae e imprime un top 10 de expleados con mayor salario incluyendo los beneficios 


```python
#todo
df.columns
df_sort = df.sort_values(by= "TotalPayBenefits", ascending= False)
top_10 = df_sort.head(10)
print(top_10[["EmployeeName", "TotalPayBenefits"]])

```

2. **Filtrar empleados con más de 50,000 en horas extra, la columna a tratar es: (OvertimePay)**

```python
#todo
# Convertir la columna 'OvertimePay' a numérico
# (puede contener caracteres no numéricos)
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce')

# Filtrar empleados con más de 50,000 en horas extra
empleados_mas_horas_extras = df[df['OvertimePay'] > 50000]

# Imprimir los resultados
print(empleados_mas_horas_extras[['EmployeeName', 'OvertimePay']])
```

3. **Contar cuántos empleados únicos hay por año**

```python
#todo
empleados_por_anio = df.groupby("Year")["EmployeeName"].nunique() # Agrupamos por año y contamos los empleados únicos
print(empleados_por_anio)
```

4. **Ver cuántos cargos únicos (JobTitle) existen y los 5 más comunes**

```python
#todo
cargos_unicos = df.groupby("JobTitle").nunique().head(5) #.nunique() cuenta cuántos valores distintos hay.
print(cargos_unicos)

cargos_comunes =  df["JobTitle"].value_counts().head(5) #.value_counts() cuenta cuántos valores hay y los ordena de mayor a menor.
print("\nCinco cargos más comunes:")
print(cargos_comunes)
```

5. **Mostrar el salario total (TotalPay) promedio por año**

```python
#todo
salario_promedio_anio = df.groupby("Year")["TotalPay"].mean() # Agrupamos por año y calculamos el salario promedio
print(salario_promedio_anio)
```
